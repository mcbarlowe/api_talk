```python
import requests
import pandas as pd
```

```python
#hitting the api endpoint with the paramenters and then parsing it from json to a python dictionary
url = 'https://stats.theseventhman.net/stats/api/v1/players/'
parameters = {'agg': 'no', 'player':'', 'team': '', 'season': '2020', 'toc':''}
response = requests.get(url, params=parameters).json()
```

```python
response
```

```python
player_df = pd.DataFrame.from_dict(response)
player_df.head()
```
